## Desicion tree using a classfication

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as  plt
import seaborn as sns
```

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

```python
print(iris['DESCR'])
# Describe statistics
```

```python
print(iris['target'])
# y traget column
```

```python
# independent features
x = pd.DataFrame(iris['data'], columns=['sepal length in cm','sepal width in cm','petal length in cm','petal width in cm'])
```

```python
x
# x  features
```

```python
# dependent variable
y = iris['target']
```

```python
y
# traget
```

```python
# train  and test
from sklearn.model_selection import  train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.80,random_state=10)
```

```python
# apply Descision  tree classifier
from sklearn.tree import DecisionTreeClassifier
treeclassifier  =  DecisionTreeClassifier()
```

```python
treeclassifier.fit(x_train,y_train)
# decision tree
```

```python
#  visualize the decision tree
from sklearn import tree
plt.figure(figsize=(10,10))
tree.plot_tree(treeclassifier,filled = True)
# using  a plot_tree
# it fill al the dat  is  filled
```

```python
y_pred = treeclassifier.predict(x_test)
# predition to assian a x_test
```

```python
from sklearn.metrics import  confusion_matrix, classification_report
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
